## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial
import numpy as np

np.random.seed(42)
print(dbinom(2,10,0.2))
print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

Write the above functions but for both normal and uniform distributions.

```python
from scipy.stats import norm

# Define all the probability functions related to this distribution

def dnormal(x, media, dst):
    """
    Calculates the point estimate of the normal distribution
    """
    result = norm.pdf(x,loc= media, scale=dst)

    return result

def pnormal(x, media, dst):
    """
    Calculates the cumulative of the normal distribution
    """
    result = norm.cdf(x,loc= media, scale=dst)

    return result

def qnormal(x, media, dst):
    """
    Calculates the quantile function from the normal distribution
    """
    result = norm.ppf(x,loc= media, scale=dst)

    return result

def rnormal(tama, media,dst):
    """
    Generates random variables from the normal distribution
    """
    result = norm.rvs(size=tama, loc=media, scale=dst)

    return result
```

```python
dnormal(1,media=2,dst=0.3)
print(f"Probability that a normal(2,0.3) takes the value 1: {dnormal(1,media=2,dst=0.3)}")
print(f"Probability that a normal(2, 0.3) will take a value less than 2: {pnormal(2,media=2,dst=0.3)}")
print(f"Which value of a normal(2, 0.3) has a cumulative probability of 0.75? {qnormal(0.75,media=2,dst= 0.3)}")
print(f"Generate 2000 random values from a normal(2, 0.3) distribution: {rnormal(tama=2000, media = 2, dst= 0.3)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

media = 2
dst = 0.3
datan = np. arange (0, 4, 0.001)

pdf = dnormal(datan, media, dst)
cdf = pnormal(datan, media, dst)
ppf = qnormal(datan, media, dst)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].plot(datan, pdf)
axis[1].plot(datan, cdf)
axis[2].plot(datan, ppf)

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
from scipy.stats import uniform

# Define all the probability functions related to this distribution

def duniform(x, min, max):
    """
    Calculates the point estimate of the uniform distribution
    """
    result = uniform.pdf(x,loc= min, scale=(max-min))

    return result

def puniform(x, min, max):
    """
    Calculates the cumulative of the uniform distribution
    """
    result = uniform.cdf(x,loc=min, scale=(max-min))

    return result

def quniform(q, min, max):
    """
    Calculates the quantile function from the uniform distribution
    """
    result = uniform.ppf(q,loc=min, scale=(max-min))

    return result

def runiform(min,max,tama):
    """
    Generates random variables from the uniform distribution
    """
    result = uniform.rvs(loc=min, scale=(max-min),size=tama)

    return result
```

```python
print(f"Probability that a uniform(0,20) takes the value 8: {duniform(8,min=0,max=20)}")
print(f"Probability that a uniform(0,20) will take a value less than 2: {puniform(2,min=0,max=20)}")
print(f"Which value of a uniform(0, 20) has a cumulative probability of 0.75? {quniform(0.75,min=0,max=20)}")
print(f"Generate 2000 random values from a uniform(0, 10) distribution: {runiform(min=0,max=10,tama=2000)}")
```

```python
np.random.seed(42)

min = 0
max = 20
datan = np. arange (-10, 30, 0.0001)

pdf = duniform(datan, min, max)
cdf = puniform(datan, min, max)
ppf = quniform(datan, min, max)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].plot(datan, pdf)
axis[1].plot(datan, cdf)
axis[2].plot(datan, ppf)

# Adjust the layout
plt.tight_layout()
```
